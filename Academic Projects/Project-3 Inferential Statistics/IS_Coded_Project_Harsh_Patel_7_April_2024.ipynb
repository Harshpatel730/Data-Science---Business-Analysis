## PROBLEM-2:

The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain.

Answer the following questions from the above information:

Q-2.1) What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?

Q-2.2) What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?

Q-2.3) What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

Q-2.4) What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?

Q-2.1) What proportion of the gunny bags have a breaking strength of less than 3.17 kg per sq cm?

```python
# We know the mean and Standard deviation as given below:
mu = 5
sigma = 1.5
```

```python
z1=(3.17-mu)/sigma
z11 = stats.norm.cdf(z1)
z11 = round(z11,4)
print(f"The proportion for having the breaking strength < 3.17 kg/sq.cm of gunny bags is {z11}. ")

x=np.linspace(0,10,100)
plt.plot(x, stats.norm.pdf(x,mu,sigma))
plt.axvline(x=3.17,c='g')
x1=np.linspace(0,3.17,20)
plt.fill_between(x1, stats.norm.pdf(x1,mu,sigma),color='g')
plt.xlabel("Gunny Bags Breaking strength(kg/sq.cm)")
plt.ylabel("Probability")
plt.title("Normal Distribution for Breaking Strength < 3.17 kg/sq.cm")
plt.show()
```

Q-2.2) What proportion of the gunny bags have a breaking strength of at least 3.6 kg per sq cm.?

```python
z2=(3.6-mu)/sigma
z22 = 1- stats.norm.cdf(z2)
z22 = round(z22,4)
print(f"The proportion for having breaking strength atleast 3.6 kg/sq.cm of gunny bags is {z22}") 
```

```python
x=np.linspace(0,10,100)
plt.plot(x, stats.norm.pdf(x,mu,sigma))
plt.axvline(x=3.6,c='g')
x1=np.linspace(3.6,x.max(),20)
plt.fill_between(x1,stats.norm.pdf(x1,mu,sigma),color='g')
plt.xlabel("Gunny Bags Breaking strength of Gunny Bags(kg/sq.cm)")
plt.ylabel("Probability")
plt.title("Normal Distribution for Breaking Strength >= 3.6 kg/sq.cm")
plt.show()
```

Q-2.3) What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

```python
z3=(5-mu)/sigma
z4=(5.5-mu)/sigma
z34 = stats.norm.cdf(z4)-stats.norm.cdf(z3)
z34 = round(z34,4)
print(f"The proportion for having breaking strength between 5 and 5.5 kg/sq.cm of gunny bags is {z34}")
```

```python
x=np.linspace(0,10,100)
plt.plot(x, stats.norm.pdf(x,mu,sigma))
plt.axvline(5,c='g')
plt.axvline(5.5,c='g')
x1=np.linspace(5,5.5,10)
plt.fill_between(x1, stats.norm.pdf(x1,mu,sigma),color='green')
plt.xlabel("Gunny Bags Breaking strength(kg/sq.cm)")
plt.ylabel("Probability")
plt.title("Normal Distribution for Breaking strength between 5 and 5.5 kg/sq.cm")
plt.show()
```

Q-2.4) What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?

```python
z5=(3-mu)/sigma
z6=(7.5-mu)/sigma
z56 = 1-(stats.norm.cdf(z6)-stats.norm.cdf(z5))
z56 = round(z56,4)
print(f"The proportion for having breaking strenght that are not betweern the 3 and 7.5 kg/sq.cm of gunny bags is {z56}")
```

```python
x=np.linspace(0,10,100)
plt.plot(x, stats.norm.pdf(x,mu,sigma))
plt.axvline(3,c='g')
plt.axvline(7.5,c='g')
x1=np.linspace(x.min(),3,20)
x2=np.linspace(7.5,x.max(),20)
plt.fill_between(x1, stats.norm.pdf(x1,mu,sigma),color='green')
plt.fill_between(x2, stats.norm.pdf(x2,mu,sigma),color='green')
plt.xlabel("Gunny Bags Breaking strength(kg/sq.cm)")
plt.ylabel("Probability")
plt.title("Normal Distribution for Breaking strength Not between 3 and 7.5 kg/sq.cm)")
plt.show()
```

Q-3.1) Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?

First we will define our null and alternative hypothesis for the given question.

NULL HYPOTHESIS : The mean Brinell's hardness for unpolished stone is greater than or equal to 150.

ALTERNATIVE HYPOTHESIS: THe mean Brinell's hardness for unpolished stone is less than 150.

Ho : µ ≥ 150

Ha : µ < 150

Here, we will consider the value for level of significance(α) = 0.05

Now, we have two samples one is 'unpolished' and another is 'treated and polished'.Also, both the sample size is same, n = 75.

we do not know the population standard deviation and from the shape function we get that the sample size is > 30, so we will use t-distribution and tSTAT test statistic.

For, this we will use ttest_samp1 function to calculate the mean of one sample.(NOTE:- This function will return t-statistic and two-tailed p-value, as for this question we are doing one tail test we will divide the p-value by 2).

```python
t_statistic, p_value = ttest_1samp(df1['Unpolished '],150)
print('tstat=',t_statistic)    
print('P Value=',p_value/2)
print("We have get the p-value < 0.05 and thus, we have enought evidence to reject the null hypothesis and consider the alternative hypothesis. The mean Brinell's Hardness for Unpolished stone is less than 150.")
```

```python
cv = stats.norm.ppf(0.05,0,1)
print(f"The critical value is {cv}.")
```

```python
x=np.linspace(-6,6,150)
plt.plot(x, stats.norm.pdf(x, 0, 1))
plt.axvline(x = t_statistic, c = 'orange',linestyle='--')
plt.axvline(x = cv, c = 'green')
x1 = np.linspace(-6,cv, 50)
plt.fill_between(x1, stats.norm.pdf(x1, 0, 1), color='r')
plt.title("t_statistic and Critical value Plot")
plt.xlabel("Orange line shows t_statistic value,Green line shows Critical value")
plt.ylabel("Probability")
plt.show()
```

As, seen from above graph, we will reject the null hypthesis as t-statistc value < critical value.

Q-3.2) Is the mean hardness of the polished and unpolished stones the same?

First of all we will define our null and alternative hypothesis for this question.

NULL HYPOTHESIS: The mean Brinell's Hardness for unpolished stone is same as mean Brinell's Hardness for treated and polished stone.

ALTERNATIVE HYPOTHESIS: The mean Brinell's Hardness for unpolished stone is not equal to mean Brinell's Hardness for treated and polished stone.

Ho: µ(unpolished) = µ(treated and polished)

Ha: µ(unpolished) ≠ µ(treated and polished)

Here also, we will consider the value for level of significance(α) = 0.05

Now, we have two samples one is 'unpolished' and another is 'treated and polished'.Also, both the sample size is same, n = 75.

we do not know the population standard deviation and from the shape function we get that the sample size is > 30, so we will use t-distribution and tSTAT test statistic.

Also, we will use the ttest_ind function for two tail t-test independent samples.

```python
t_statistic, p_value  = stats.ttest_ind(df1['Unpolished '],df1['Treated and Polished'])
print('tstat=',t_statistic)    
print('P Value=',p_value)
print("Since, the value of p-value is < alpha(0.05), we will reject the null hypothesis and consider the alternative hypthesis. The mean hardness for unpolished stone is not same as treated and polished stone.")
```

```python
cv1 = stats.norm.ppf(1-(0.05/2))
cv2 = stats.norm.ppf(0.05/2)
print(f"The critical value-1 is {cv1}.\nThe critical value 2 is {cv2}.")
```

```python
x = np.linspace(-4,4,150)
plt.plot(x, stats.norm.pdf(x, 0, 1))
plt.axvline(t_statistic,c='orange',linestyle = '--')
plt.axvline(x = cv1, c = 'g')
x1 = np.linspace(cv1, 4, 50) # 
plt.fill_between(x1, stats.norm.pdf(x1, 0, 1), color='r')
plt.axvline(x = cv2, c = 'g')
x2 = np.linspace(-4, cv2, 50) 
plt.fill_between(x2, stats.norm.pdf(x2, 0, 1),color='r')
plt.annotate('Reject-Null', (2.2, 0.20)) 
plt.annotate('Reject-Null', (-3.5, 0.20)) 
plt.annotate('  Do Not Reject\n       Null', (-1, 0.20)) 
plt.title("Showing t_statistic value present in Rejection Region")
plt.xlabel("Yellow line shows t_statistic Value,Green line shows Critical value")
plt.ylabel("Probabiliy")
plt.show() 
```

As seen from the above graph we will reject the null hypothesis the t-stats value is in rejection zones.

Q-4.1) How does the hardness of implants vary depending on dentists?

1. Null and Alternative Hypothesis for Alloy-1:
   
   H01: Mean Hardness is same across all dentist for alloy-1.
   
   Ha1: Mean Hardness is not same for atleast one pair of dentist for alloy-1.
   ####
2. Null and Alternative Hypothesis for Alloy-2:
   
   H02: Mean Hardness is same across all dentist for alloy-2.
   
   Ha2: Mean Hardness is not same for atleast one pair of dentist for alloy-2.

```python
formula = 'Response ~ C(Dentist)'
model = ols(formula,Alloy1).fit()
aov_table = anova_lm(model)
print(aov_table)
```

Thus, P-value is > 0.05(alpha), we fail to reject the null hypothesis and we can say that, for alloy-1 the implant hardness is same for all dentist.

```python
plt.title("Dentist vs Response Point Plot for Alloy-1")
sns.pointplot(Alloy1,x='Dentist',y='Response',ci=None)
```

Even though we failed to reject the null hypothesis we will still use Tukey Hsd test to find out the differnt part for alloy-1.

```python
Multic = pairwise_tukeyhsd(Alloy1.Response,Alloy1.Dentist,alpha=0.05)
print(Multic)
```

Since, we fail to reject the Null hypothesis it is not possible to identify which pair of dentist are different as mean hardness for implant is same for all dentist for alloy-1

```python
formula = 'Response ~ C(Dentist)'
model = ols(formula,Alloy2).fit()
aov_table = anova_lm(model)
print(aov_table)
```
